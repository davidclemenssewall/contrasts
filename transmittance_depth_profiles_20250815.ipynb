```python
# imports
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Read in a1 data
data_path = os.path.join('..', 'data', 'whoi_l_arm_transmittance')

project_names = ['20250730-r2-ssl',
                 '20250730-r2-NEpond',
                 '20250730-r2-Spond',
                 ]

ls_df_a1 = []
for project_name in project_names:
    file_path = os.path.join(data_path,
                            project_name,
                            'contrasts_l_arm_' + project_name + '.a1.h5')
    ls_df_a1.append(pd.read_hdf(file_path))

df_a1 = pd.concat(ls_df_a1)
```

```python
df_a1
```

```python
# Find depth profile
# NEpond
location = 'NEpond'
azimuth = '235'

df_temp = df_a1.query('location == @location & azimuth == @azimuth & type == "T" & variable == "Pressure"').reset_index()

plt.scatter(x='timestamp_utc',y='value', data=df_temp)

start = '2025-07-30 15:47:25+00:00 	'
end = '2025-07-30 15:50:40+00:00'

plt.axvline(pd.Timestamp(start))
plt.axvline(pd.Timestamp(end))
```

```python
df_temp
```

```python
# Combine inclination, pressure, and depth values
location = 'NEpond'
azimuth = '235'
start = '2025-07-30 15:47:29+00:00'
end = '2025-07-30 15:50:40+00:00'

depth_gauge = np.arange(16)*0.1 + 1.5
depth_gauge[0] = 1.55
sensor_offset = 0.80
ice_draft = 0.75

df_temp_1 = df_a1.query('timestamp_utc >= @start & azimuth == @azimuth & location == @location & type == "T" & variable == "Transmittance"')
df_temp_2 = df_a1.query('timestamp_utc >= @start & azimuth == @azimuth & location == @location & type == "T" & variable in ["InclV", "Pressure"]')

df_temp_2 = df_temp_2.reset_index().drop(columns=['station', 'location', 'azimuth', 'type', 'depth', 'repetition', 'wavelength_nm'])
df_temp_2 = df_temp_2.pivot(index='timestamp_utc', columns='variable', values='value').reset_index()
df_temp_2['depth_gauge'] = depth_gauge
df_temp_2['depth_water'] = df_temp_2['depth_gauge'] - sensor_offset
df_temp_2['depth_ice'] = df_temp_2['depth_water'] - ice_draft

df_temp = df_temp_1.reset_index().merge(df_temp_2, left_on='timestamp_utc', right_on='timestamp_utc')
```

```python
wv_min = 380
wv_max = 750

f, axs = plt.subplots(2, 1, figsize=(5, 6))

sns.lineplot(x='wavelength_nm', y='value', hue='depth_ice', data=df_temp, ax=axs[0])
axs[0].grid()

sns.lineplot(x='depth_ice', y='value', hue='wavelength_nm', ax=axs[1], palette='rainbow', alpha=0.3,
            data=df_temp.query('wavelength_nm >= @wv_min & wavelength_nm <= @wv_max'))
```
