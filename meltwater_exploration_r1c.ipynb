# meltwater exploration

```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = os.path.join('..', 'data', 'whoi_ysi')
filename = 'station1X_meltwater_scratchwork.xlsx'
```

```python
df = pd.read_excel(os.path.join(data_path, filename))
```

```python
df
```

```python
df_temp = df.query('line == "lead"')
df_temp['distance_from_lead'] = df_temp['hole'].astype('float')

df_temp2 = df.query('line == "ROV_S"')
df_temp2['distance_from_lead'] = df_temp2['hole'].astype('float')
df_temp2['distance_from_lead'] -= 99.0
df_temp2['distance_from_lead'] *= -1
                   
df_temp = pd.concat([df_temp, df_temp2])
```

```python
df_temp
```

```python
f, axs = plt.subplots(2, 1, sharex=True, figsize=(6, 6))

sns.lineplot(x='distance_from_lead', y='draft', hue='line', data=df_temp, ax=axs[0])

axs[0].set_ylim([80, 0])
axs[0].set_ylabel('Draft (cm)')

sns.lineplot(x='distance_from_lead', y='mw_thickness', hue='line', data=df_temp, ax=axs[1])

axs[1].set_ylim([12, 0])

axs[1].set_ylabel('Meltwater thickness (cm)')
axs[1].set_xlabel('Distance from lead (m)')

f.savefig(os.path.join('..', 'figures', '4.YSI.1_meltwater_lens_transects_1c.png'),
         dpi=300, bbox_inches='tight')
```
